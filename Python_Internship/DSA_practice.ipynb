```python
# Approach 1: Shift from the back
def insert (arr, x):
  if x <= 0:
    return False
  if arr[-1] != 0:
    return False
  i = len(arr) - 2
  while i >= 0 and x < arr[i]:
    arr[i+1] = arr[i]
    i = i - 1
  arr[i+1] = x
  return True
```

```python
# Approach 2: Swap while moving forward

def insert_swap(arr, x):
  if x <= 0:
    return False
  if arr[-1] != 0:
    return False
  for i in range(len(arr)):
    if x < arr[i] or arr[i] == 0:
      temp = arr[i]
      arr[i] = x
      x = temp
  return True

```

```python
categories = ['shoes', 'shirts', 'shoes', 'jackets', 'shirts', 'shoes']

counts = {}

for i in categories:
  if i in counts:
    counts[i] += 1
  else:
    counts[i] = 1

print(counts)

top_category_1 = max(counts, key=counts.get)
print(top_category_1)
```

### Exploring Built-in Functions vs. Methods

A **function** is a piece of code that is called by name. It can be passed data to operate on (i.e. the parameters) and can optionally return data (the return value).

A **method** is a piece of code that is called by a name that is associated with an object. In most respects it is identical to a function except for two key differences:
1. It is implicitly passed the object on which it was called.
2. It is able to operate on data that is contained within the class (the object is an instance of the class).

```python
import builtins

# Get all attributes of the builtins module
all_builtins = dir(builtins)

# Filter to find only the built-in functions (starting with lowercase)
# We exclude objects that start with Uppercase (usually Exceptions like ValueError)
builtin_functions = [item for item in all_builtins if item[0].islower() and not item.startswith('__')]

print("Common Built-in Functions:")
print(builtin_functions)
```

### Understanding Classes and Objects

To truly understand OOP, let's create our own 'Blueprint' (Class) for a `Dog`.

```python
class Dog:
    # The __init__ method sets up the 'attributes' (data) for each object
    def __init__(self, name, breed):
        self.name = name   # Attribute
        self.breed = breed # Attribute

    # A 'method' is a function that belongs to this class
    def bark(self):
        return f"{self.name} says Woof!"

# --- Creating OBJECTS from the Class ---

# 'dog1' is an Object (an instance of the Dog class)
dog1 = Dog("Buddy", "Golden Retriever")

# 'dog2' is another Object
dog2 = Dog("Luna", "Labrador")

print(f"Dog 1: {dog1.name}, a {dog1.breed}")
print(dog1.bark())

print(f"\nDog 2: {dog2.name}, a {dog2.breed}")
print(dog2.bark())
```

### Why do we need `self`?

Imagine we have two dogs. When we call `.bark()`, `self` is what tells Python which dog's name to use. It's essentially a placeholder for the specific object.

```python
# Let's look at it in action
print(f"When we call dog1.bark(), 'self' inside the method becomes dog1 (Buddy).")
print(f"Result: {dog1.bark()}")

print(f"\nWhen we call dog2.bark(), 'self' inside the method becomes dog2 (Luna).")
print(f"Result: {dog2.bark()}")
```

### The DSA Analogy: Pantry vs. Recipe

To see how Algorithms and Data Structures work together in OOP, let's build a Cake Baking class.

```python
class CakeRecipe:
    def __init__(self, flavor, ingredients):
        # DATA STRUCTURE: This dictionary is our 'Pantry'
        self.flavor = flavor
        self.pantry = ingredients

    # ALGORITHM: The step-by-step instructions
    def bake(self):
        print(f"--- Starting the {self.flavor} Cake Algorithm ---")

        # Step 1: Check data
        for item, amount in self.pantry.items():
            print(f"Mixing {amount} of {item}...")

        # Step 2: Process data
        print("Heating oven to 350 degrees...")

        # Step 3: Return result
        return f"Successfully baked a delicious {self.flavor} cake!"

# --- Creating the Object ---
my_ingredients = {"flour": "2 cups", "sugar": "1 cup", "eggs": 2}
chocolate_cake = CakeRecipe("Chocolate", my_ingredients)

# Running the Algorithm
result = chocolate_cake.bake()
print(f"\nOutcome: {result}")
```

### Inspecting the Object's Data

Let's look inside the `chocolate_cake` object to see how it stored our information.

```python
# Show the 'flavor' attribute (which is a String)
print(f"Flavor Type: {type(chocolate_cake.flavor)}")
print(f"Flavor Value: {chocolate_cake.flavor}")

print("\n" + "-"*20 + "\n")

# Show the 'pantry' attribute (which is a Dictionary)
print(f"Pantry Type: {type(chocolate_cake.pantry)}")
print("Pantry Contents:")
display(chocolate_cake.pantry)
```
