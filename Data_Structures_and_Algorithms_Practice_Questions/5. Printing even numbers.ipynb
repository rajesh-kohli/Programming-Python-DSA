#### Given a positive number n , design an algorithm to print all the even numbers from 2 to n

```python
n = int(input("Enter any natural number: "))
print(n)
```

```python
i = 2
while i<=n and i>=2:
    if i%2 == 0:
        print(i)
    i +=1
```

```python
for i in range(2,n+1,1):
    if i%2 == 0:
        print(i)
```
