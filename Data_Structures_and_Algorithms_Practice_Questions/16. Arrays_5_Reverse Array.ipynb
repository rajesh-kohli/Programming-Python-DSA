## Given an array of N integers, design an algorithm to reverse it

```python
# time: n/2.constant ~ O(n)
def reverse(arr: list[int], n: int) -> None:
    i, j = 0, n-1
    while i < j:
        arr[i], arr[j] = arr[j], arr[i]
        i += 1
        j -= 1

n = int(input())
arr = [int(input()) for _ in range(n)]
print(arr) #
reverse(arr, n)
print(arr)
```
