```python
def is_valid_parentheses(s: str) -> bool:
    # Map each closing bracket to its corresponding opening bracket
    bracket_map = {')': '(', '}': '{', ']': '['}
    stack = []
    
    for char in s:
        # If it is a closing bracket
        if char in bracket_map:
            # Pop the top element if stack isn't empty, else use a dummy value
            top_element = stack.pop() if stack else '#'
            
            # If the popped element doesn't match the required opening bracket
            if bracket_map[char] != top_element:
                return False
        else:
            # It is an opening bracket, push it onto the stack
            stack.append(char)
            
    # If the stack is empty, all brackets were matched correctly
    return len(stack) == 0

# --- Test Cases ---
print(is_valid_parentheses("()[]{}"))  # Output: True
print(is_valid_parentheses("([)]"))    # Output: False (Incorrect nesting)
print(is_valid_parentheses("("))       # Output: False (Unclosed)
```

```python
def check_parentheses(s: str) -> str:
    stack = []
    
    for char in s:
        if char == '(':
            stack.append(char)
        elif char == ')':
            # Mismatch: A closing bracket appears but stack is empty
            if not stack:
                return "not match"
            stack.pop()
            
    # After checking all characters:
    if len(stack) > 0:
        # Elements are left over, meaning the user left the input hanging
        return "incompleteInputError: incomplete input"
        
    return "match"

# --- Test Cases ---
print(check_parentheses("(())"))  # Output: match
print(check_parentheses("(()"))   # Output: incompleteInputError: incomplete input
print(check_parentheses("())"))   # Output: not match
print(check_parentheses(")("))    # Output: not match
```
