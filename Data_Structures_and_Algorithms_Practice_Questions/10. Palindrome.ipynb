```python
s = str(input("Enter a string: "))
```

```python
def is_palindrome(s):
    if s == s[::-1]:
        print (f"The string {s} is a Palindrome")
    else:
        print ("Not a Palindrome")
```

```python
## Boolean way

def palindrome_check(s):
    return s == s[::-1]
```

```python
is_palindrome(s)
```

```python
test1 = "racecar"
is_palindrome(test1)
```

```python
palindrome_check(test1)
```
