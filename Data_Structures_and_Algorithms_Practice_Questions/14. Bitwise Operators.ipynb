### Bitwise Operators

1. Bitwise AND (&) Operator

```python
a = 25
b = 19
print(a & b)

## Binary of 25: 1 1 0 0 1 (Bit 4,3 and 0 are set)
## Binary of 19: 1 0 0 1 1 (Bit 4,1 and 0 are set)
## 16 + 1 = 17 (Because Bit 4(2^4 = 16) and Bit 0 (2^0 = 1) are 1, 
#               others are 0)
```

```python
## OR represented by |
print(a|b)
## Bits 4,3,1,and 0 are set in the OR operation
## 2^4 + 2^3 + 2^1 + 2^0 = 16 + 8 + 2 + 1 = 27

```

## a&b will always be <= min(a,b) - Result of AND will always be <= min(a and b)
## a|b will always be >= max(a,b) 

```python
## XOR represented by  ^
print(a ^ b)
```

```python
## Another trick about XOR is - if the number of set-bits for a given bit position is odd, 
# it's XOR would be 1 for that position and if it's even, 
# it would be 0 for that bit position
# This is helpful when we are doing XOR for more than 2 numbers
```

```python
## set Kth bit
## Given a non negative integer N, design an algorithm to set it's Kth bit
# Input : N = 42, K = 4 
# Output : 58 

n = int(input())
k = int(input())

mask = 1 << k
ans = n | mask

print(ans)

# mask is a number whose kth bit is set, rest all the bits are zero. 
# that would be 1<<k

# So we do an OR of n and mask to get the solution
```

```python
## clear Kth bit

## here I want all the bits 1 and the kth bit 0 
# and then I do the AND to get the desired number

## Here Mask is a number where all bits are 1 except kth bit
# for that we just take a complement of ~(1<<k)

n = int(input())
k = int(input())

mask = ~(1 << k)
ans = n & mask

print(ans)
```

```python
## check if kth bit is set or not

## n >> k - when you do this,
# kth bit of n would become 0th bit in the resultant value

## to solve this question - we can check if the 0th bit of the resultant value is set or not
# meaning: (n>>k) & 1 --> if 0 (oth bit of (n>>k) is 0 which means kth bit of n is 0)
#                     --> if 1 (oth bit of (n>>k) is 1 which means kth bit of n is 1)

n = int(input())
k = int(input())

if (n >> k) & 1:
    #print("kth bit of n is set")
    print(True)
else:
    #print("kth bit of n is not set")
    print(False)
```

```python
## Flip kth bit
n = int(input())
k = int(input())

mask = 1 << k
ans = n ^ mask
```

```python
### count set bits
n = int(input())
cnt = 0
 
while n != 0:
     
    zerothBit = n&1
    if zerothBit == 1:
        cnt = cnt + 1
    n =  n >> 1
print(cnt)
```

```python
## simplified version
n = int(input())
cnt = 0

## iterations = ceil(log2(n+1)) ~ logn
# work done in each iteration = constant
# time: O(logn)

while n != 0:
    cnt = cnt + (n & 1)
    n = n >> 1
print(cnt)
```

```python
## Even better strategy for counting the number of set bits
# Brian Kernighan's Algorithm

## it tells you to only iterate on the set bits, and not iterate on 0's along with 1's
# hot to write code to skip to set-bits directly and not go through each one by one

n = int(input())
cnt = 0

# iteration = no of set bits in n
# work done in each iteration is constant
# time: O (no. of set bits in n)

while n != 0:
    
    # turn off the right most set bit of n
    n = n & (n - 1)
    cnt = cnt + 1

print(cnt)
```
