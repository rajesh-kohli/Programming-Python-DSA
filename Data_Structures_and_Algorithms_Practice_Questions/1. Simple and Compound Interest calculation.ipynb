#### calculate compound interest using this mathematical formula:

##### two different ways of writing the equation in LaTEX code
$$ A = P \left(1 + \frac{r}{n}\right)^{nt} $$

$A = P(1+ r/n)^{nt}$

- **A**: Total final amount *(Principal + Interest)*
- **P**: Principal amount
- **r**: Annual interest rate *(as a decimal, e.g., 7% = 0.07)*
- **t**: Time in years
- **n**: Number of times interest compounds per year *(use 12 for monthly compounding)*

> **Important Note:** Standard financial formulas expect the interest rate `r` to be a decimal (e.g., 0.07 instead of 7).

***Syntax Breakdown***  (Centered (Display) Mode)
- **\left( and \right)**: Automatically resizes the parentheses so they don't look too small next to a tall fraction.
- **\frac{numerator}{denominator}**: Creates a vertical fraction.
- **^{nt}**: The ^ makes text a superscript (exponent). Wrapping nt in {} ensures both letters stay up in the exponent together.

**Inline Mode** To fit a formula right inside a sentence (like \(A = P\)), use single dollar signs

```python
P, r, t, n = map(int, input("Enter Principal, rate of interest in percent, time in years, number of times interest compounds per year").split())
```

```python
P, r, t, n
```

```python
## Compound_Interest

Total_Final_Amount = P * (1 + (r/100)/n) **( n*t)
print(Total_Final_Amount)
```

```python
Total_Final_Amount = round(P * (1 + (r/100)/n) **( n*t))
print(Total_Final_Amount)
```
