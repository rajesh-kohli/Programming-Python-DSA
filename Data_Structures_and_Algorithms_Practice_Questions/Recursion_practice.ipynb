#### Divide and Conquer Strategy - This is a Recursive Algorithm - Implemented for Recursion only
- Let's say we have a problem P, assume that problem is large problem
- if the problem is large, divide that problem into subproblem and continue dividing it, until it becomes small enough
- When do we know it's small enough
    - if the problem is getting solved using one or two operations, then it's small and we stop
- And then you combine all the solutions of the small problems to get a final solution

##### When to use Divide and Conquer strategy
- let's say the larger problem is sorting problem and if we divide that into small problems, those small problems should be sorting problems
- Meaning the main problem and sub-problems should be the same problem statements

###### How to Divide:
- E.g. MIN-MAX Problem

```python
def Recursive_min_max(lst, low, high):
    ## if there is single element in a list, both mix and max are the same
    ## so the index 0 is both low and high
    ## if there are two elements, the index 0 is low and index 1 is high
    ## And these two are my anchor steps
    if low == high: ## single element list
        return lst[low], lst[high]
    elif high == low + 1:
        if lst[high] > lst[low]:
            return lst[low], lst[high]
        else:
            return lst[high], lst[low]
    else:
        mid = (low + high)//2
        left_min, left_max = Recursive_min_max(lst, low, mid)
        right_min, right_max = Recursive_min_max(lst, mid+1, high)
        original_min = minimum(left_min, right_min)
        original_max = maximum(left_max, right_max)
        return original_min, original_max

def minimum(a,b):
    if a < b:
        return a
    else:
        return b

def maximum(a,b):
    if a > b:
        return a
    else:
        return b
```

#### Problem statement: Find the power of a number using divide and conquer strategy
##### power(a,n)

```python
def recursive_power(a,n):
    if n == 0:
        return 1
    elif n%2 == 0:
        res1 = recursive_power(a,n//2)
        return res1*res1
    else:
        res2 = recursive_power(a,n//2)
        return res2*res2*a
```

```python
## Given two sorted lists, append the elements in the new list (merge the two list) - extend function

def merging(lst1, lst2):
    i = 0
    j = 0
    result = []
    while i < len(lst1) and j<len(lst2):
        if lst1[i] < lst2[j]:
            result.append(lst1[i])
            i+=1
        else:
            result.append(lst2[j])
            j+=1
    if i==len(lst1):
        while j<len(lst2):
            result.append(lst2[j])
            j+=1
    if j == len(lst2):
        while i<len(lst1):
            result.append(lst1[i])
            i+=1
    return result
```

#### Two pointer concept (in DSA/Competitve programming)
- keep swapping the two pointers at first and last indexe
- once the two pointers are at same index, you stop


```python
def my_reverse_iteration(lst):
    i = 0
    j = len(lst) - 1
    while i < j:
        lst[i],lst[j] = lst[j], lst[i]
        i += 1
        j -= 1
    return lst
```

```python
def recursion_variation_reverse(lst, i, j):
    if i > j:
        return
    lst[i], lst[j] = lst[j],lst[i]
    recursion_variation_reverse(lst, i+1, j-1)
    return lst
```

```python
### Identify common element in two lists and find the intersection and store in dictionary
## this dictionary approach is the fastest - O(1)

def common_elements(lst1, lst2):
    result = []
    smallest = {}
    for i in lst2:
        smallest[i] = 1
    for i in lst1:
        if i in smallest:
            result.append(i)
    return result
```

```python
lst1 = [3,4,5,6,7,8,9,15]
lst2 = [8,9,10,15]
common_elements(lst1, lst2)
```
