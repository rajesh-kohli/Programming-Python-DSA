### Graph
 - we use dictionary data structure to construct a graph
 - A graph is a non-linear data structure, denoted by G(V,E)
    - V --> no of vertices
    - E --> no of Edges
        - Example: V = {A,B,C,D}
        - E = {A-D, A-B, B-C, C-D}
    - A graph is used to show the relationship between the entities
    - Very common Data Structure when it comes to real-world problems representation

    - Eg. Person drives a car
        - entity 1 --> person (represented by one Vertex)
        - entity 2 --> car (represented by another vertex)
        - driving --> relationship between entity 1 and entity 2 (represented by edge/line connecting person and car)
    - Other examples: Instagram, Twitter, Facebook

- Two types of graphs:
    - Directed graph --> direction is specified (eg A->B direction is specified, the reverse isn't possible)
        - Instagram is directed graph
        - If one person follows another person, that doesn't mean both follow each other, only one person sees the other person story. Both have to follow each other to see each other content.
    - Undirected graph --> direction is not specified (eg A-B, B-A mean the same)
        - in twitter if you follow someone, whatever you are doing, both can see each other content. Or evenif you don't follow each other, you still can see. No need tp establish the relationship or follow each other unlike Instagram or Facebook if your profile is private.
    
    


```python


## We have a square ABCD (graph) with 4 vertices A, B, C, D 
# and we need to represent relationships of all 4 vertices using edges 
# and add them in a list 
# which would be values of the dictionary for each key/vertex
# for key A - there would be values in alist [B,D]
# for key B - there would be values in list [A,C] and so on

# first create a function to 

def Creation_graph(node1, node2):
    graph = {} # Empty dictionary
    if node1 not in graph:
        graph[node1] = []
    if node2 not in graph:
        graph[node2] = []
    graph[node1].append(node2)
    graph[node2].append(node1)
    return graph
    
```

```python
Creation_graph("A","B")
```

```python
graph = {}
def add_nodes(node):
    if node not in graph:
        graph[node] = []
    return graph

def add_edge(n1,n2):
    graph[n1].append(n2)
    graph[n2].append(n1)
    return graph
```

```python
add_nodes("A")
add_nodes("B")
add_nodes("C")
add_nodes("D")
```

```python
add_edge("A","B")
```

```python
add_edge("B","D")
```

```python
add_edge("A","C")
```
