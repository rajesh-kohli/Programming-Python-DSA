## An array is a linear data structure (think of data structure like a container)
## referred by a single name, that is used to store a sequence of values in a contiguous manner

## In Python, two implementations of arrays are: Lists (mutable) and Tuple (immutable)

```python
arr = [1,2,3,4,5]

print(arr[0], arr[-5])
print(arr[4], arr[-1])

print(len(arr)) ## len function
```

```python
arr = list("coding") ## converts a string object to a list of characters object
print(arr)
```

```python
## to create an empty list

arr = list()

# OR

arr = []
```

```python
nums = [10, 20, 30, 40, 50]

i = 0
n = len(nums)

while i < n:
    print(nums[i])
    i += 1
```

```python
## with for loop, you can very easily iterate over list / tuple compared to while loop
## no need for loop variable, condition, increment etc. unlike while loop
nums = [10, 20, 30, 40, 50]
for num in nums:
    print(num)
```

```python
## if we were to use for loop and need to use index, we can use range function
for i in range(len(nums)):
    print(nums[i])
```

```python
## to get both the index and the value, we can use ennumerate function

for i, num in enumerate(nums):
    print(i, num)
```

```python
## basic operations on a list object

nums = [10, 20, 30, 40, 50]
t = int(input())

for num in nums:
    if num == t:
        print(True)
        break
else:
    print(False)
```

```python
## Another way

flag = False # assume t is not present in nums

for num in nums:
    if num == t:
        print(True)
        flag = True
        break
if flag == False:
    print(flag)
```

```python
print(t in nums) ## better way (membership test)
```

```python
## compute sum of elements - instead of brute force way, use sum() function in Python

nums = [10, 20, 30, 40, 50]

## Brute force

s = 0
for num in nums:
    s += num

print(s)
```

```python
## sum function
print(sum(nums))
```

```python
## built min and max functions
print(min(nums))
 
print(max(nums))
```

```python
max_so_far = float('-inf')# initialize it with something very very small - negative infinity
for num in nums:
    if num > max_so_far:
        max_so_far = num

print(max_so_far)
```

```python
min_so_far = float('inf')# initialize it with something very very Big - positive infinity
for num in nums:
    if num < min_so_far:
        min_so_far = num

print(min_so_far)
```

```python
## slicing a list
## if you want to extract sub-array, we can use slicing concept

arr = [10,20,30,40,50]

print(arr[1:4])
```
