Given a sequence of N non-negative numbers where every number is present twice except for one number which is present only once. Design an algorithm to find the unique number without using any extra space. 


Example:


Input : N = 7

3 1 4 1 2 3 4


Output : 2

```python
## a ^ a = 0
## a ^ 0 = a

n = int(input())
ans = 0
for _ in range(n):
    x = int(input())
    ans = ans ^ x
print(ans)
```
