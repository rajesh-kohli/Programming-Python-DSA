```python
### write a function to check if a given integer is even

def isEven(n):
    if n%2 == 0:
        print("Even")
    else:
        print("odd")
```

```python
## A function to increment an integer

def incr(x:int) -> None:
    x = x + 1

a = 10
print("before incr()",a)
incr(a)
print("after incr()", a) ## remember integer obkects are immutable, and after increment, 
# a still points to 10 and a new variable gets created for incrememnt object
```

```python
def incr(x:int) -> None:
    x = x + 1
    return x

a = 10
incr(a)
```

```python
## A function to check whether an integer num is prime

def isPrime(n:int)->bool:
    i = 2
    while i * i <= n:
        if n%i == 0:
            return False
        i = i + 1
    return True


num = int(input())
print(isPrime(num))
```

```python
## A function to print first m prime numbers

# if m = 5, output should be first prime numbers: 2,3,5,7,11

def printPrimes(m:int) -> None:
    num = 2
    cnt = 0
    
    while cnt < m:
        if (isPrime(num)): ## reusing the function from the last question (function inside a function)
            print(num)
            cnt += 1
        num += 1
        
        
m = int(input())
printPrimes(m)
```
