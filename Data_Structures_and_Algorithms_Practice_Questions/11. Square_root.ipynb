```python
# Given n, find the square root

n = int(input())
ans = 0

while ans*ans <=n:
    ans = ans + 1

ans = ans - 1
print(ans)
```

```python
## the above code won't perfectly work for numbers whose integer square root doesn't exist like 30
n = int(input())
ans = 0

while ans*ans <=n:
    ans = ans + 0.1

ans = ans - 0.1

while ans*ans <=n:
    ans = ans + 0.01

ans = ans - 0.01

while ans*ans <=n:
    ans = ans + 0.001

ans = ans - 0.001

print(ans)
```

```python
## the above code won't perfectly work for numbers whose integer square root doesn't exist like 30
n = int(input())
p = int(input())
ans = 0

while ans*ans <=n:
    ans = ans + 0.01 ## if I want the answer to 2 decimal places, use 0.01, three decimal places 0.001

ans = ans - 0.01
print(ans)
```

```python
n = int(input())
p = int(input())

ans = 0

while ans * ans <= n:
    ans = ans + 1
ans = ans - 1

inc_fac = 0.1

for _ in range(p):
    while ans * ans <= n:
        ans = ans + inc_fac    
    ans = ans - inc_fac
    inc_fac = inc_fac / 10
    
print(round(ans, p))

print(f"{ans:.4f}")
```
