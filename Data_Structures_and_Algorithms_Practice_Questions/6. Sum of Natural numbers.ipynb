#### Given a positive number n , design an algorithm to compute the sum of first n natural numbers

```python
n = int(input("Enter any natural number: "))
print(n)
sum_n = 0
```

```python
for i in range(1,n+1,1):
    sum_n = sum_n + i
print(sum_n)
```

```python
i = 1
sum_n1 = 0
while i > 0 and i <=n:
    sum_n1 = sum_n1 + i
    i +=1
print(sum_n1)
```
