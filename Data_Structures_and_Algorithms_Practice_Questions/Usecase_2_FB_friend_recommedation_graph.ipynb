### Facebook friend recommendation - Graph

```python
graph = {}
def add_node(node):
    if node not in graph:
        graph[node] = []


def add_edge(node1,node2):
    graph[node1].append(node2)
    graph[node2].append(node1)
    return graph
    
            
```

```python
add_node("mayur")
```

```python
add_node("sakshi")
```

```python
add_node("rahul")
```

```python
add_node("dev")
```

```python
add_edge("mayur","sakshi")
```

```python
add_edge("sakshi","dev")
```

```python
add_edge("dev","rahul")
```

```python
#recommendation = []
def friend_recommendation(person):   
    if person in graph:
        # recommendation = []
        for friend in graph[person]:  ## iterate on the list
            recommendation = []
            if friend != person and friend not in graph[person]:
                recommendation.append(friend)
    return recommendation
```

```python
friend_recommendation("mayur")
```
