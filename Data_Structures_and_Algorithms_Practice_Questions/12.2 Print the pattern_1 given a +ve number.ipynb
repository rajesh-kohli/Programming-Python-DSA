```python
n = int(input())
i = 1

while i <= n:
    # for the ith row, print i stars
    j = 1
    while j <= i:
        print(j, end = " ") # instead of going to the next line after print
        j = j + 1
    i = i + 1
    print() ## for new line
```

```python
# can also do this with for loop
n = int(input())
i = 1
for i in range(1,n+1):
    for j in range(1,i+1):
        print(j, end = " ")
    print()
```

```python
n = int(input())
i = 1

while i <= n:
    # for the ith row, print i numbers starting from 1
    # here we want to use i and j only as counters and not use it to print, so we are using num
    j = 1
    num = 1
    while j <= i:
        print(num, end = " ") # instead of going to the next line after print
        j = j + 1
        num = num + 1
    i = i + 1
    print() ## for new line
```
