#### Given a positive number n followed by n integer values design an algorithm to find the sum of  n numbers
Example:

Input : n = 5 ; [ 10, 20, 30, 40, 50 ]

Output : 150

Explanation : 10+20+30+40+50 = 150

```python
n = int(input("Enter any postive number: "))
print(n)
```

```python
lst = list(map(int,input(f"Enter any {n} integer values: ").split()))
print(lst)
```

```python
sum_1 = 0
for i in range(n):
    sum_1 = sum_1 + lst[i]
print(sum_1)
```
