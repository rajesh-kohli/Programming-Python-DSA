#### Given a positive number n (>1), design an algorithm to check if the given number is a prime number or not 

```python
n = int(input("Enter any postive number: "))
print(n)
```

```python
import math

if n<=1:
    print("Number is not Prime")
else:
    for i in range(2,int(math.sqrt(n)+1),1):
        if n%i == 0:
            print("Not Prime")
            break
    else:
        print("Prime number")
```
