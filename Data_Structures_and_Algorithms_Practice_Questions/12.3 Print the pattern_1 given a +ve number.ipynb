```python
n = int(input())
i = 1
num = 1

while i <= n:
    # for the ith row, print i numbers starting from 1
    # here we want to use i and j only as counters and not use it to print, so we are using num
    j = 1
    
    while j <= i:
        print(num, end = " ") # instead of going to the next line after print
        j = j + 1
        num = num + 1
    i = i + 1
    print() ## for new line
```
