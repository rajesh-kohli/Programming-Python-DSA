```python
n = int(input())

for i in range (1, n+1):
    # for the ith row print n-i+1 characters starting with A in increasing order
    ch = "A"
    for _ in range(n - i + 1):
        print(ch, end = " ")
        ch = chr(ord(ch) + 1) ## need to do ASCII conversion first. Directly n+1 won't work for characters
    ## followed by n-i+1 characters in decreasing order
    
    ch = chr(ord(ch) - 1) ## Instead of E, it would start from F if we don't do this before entering in the next inner loop. because ch +1 in the above loop in the last iteration, bring it back to ch
    
    for _ in range( n - i + 1):
        print(ch, end = " ")
        ch = chr(ord(ch) - 1)
    print()
```
