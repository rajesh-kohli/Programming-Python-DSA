```python
## Alternatively comma separator:
x, y, z = map(int, input(" Enter the numbers x, y and z separated by , : ").split(","))
print(f"(x:{x}, y:{y}, z:{z})")
```

```python
## space separator - default:
d, e, f = map(int, input(" Enter the numbers d, e and f : ").split())
print(f"(x:{d}, y:{e}, z:{f})")
```

```python
# Saving an unknown number of integers straight into a list
numbers = list(map(int, input("Enter as many numbers as you want: ").split()))
print(numbers)
```

```python
if a>b and a>c:
    print(f"a: {a} is the largest among the three numbers")
elif b>c:
    print(f"b: {b} is the largest among the three numbers")
else:
    print((f"c: {c} is the largest among the three numbers"))
```

```python
## Largest of n numbers
n = numbers[0] # using n = 0 restricts it to postive numbers
# No need for i = 0 outside the loop: The for loop automatically initializes i to 0.
for i in range((len(numbers))):
    if numbers[i] > n:
        n = numbers[i]
    # i = i + 1 ## The for loop automatically increments i at each step. 
                # Manually increasing it does nothing because Python overwrites it on the next iteration.
print(n)
```

```python
# I can also Loop directly through the numbers and not use range function
# Also no need for indexing through the list when looping directly

largest = numbers[0]

for i in numbers:
    if i > largest:               #numbers[i] > largest won't work in this case, as looping directly through the list.
        largest = i
print(largest)
    
```
