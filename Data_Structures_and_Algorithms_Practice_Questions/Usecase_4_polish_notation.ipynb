### Stack: Polish Notation

a + b * c + d / f ----> prefix --> * , /

a + (*bc) + d/f

a + (*bc) + (/df)

(+*abc) + (/df)

++*abc+/df --> prefix

e.g. 2 3 * 4 + --> solve this expression using stack

--> whenever there is operand (numbers) you push it onto the stack
--> And if there is operator, apply that operation on the elements on the stack - starting with the element in the base and move to top

1. Operand --> push onto the stack
2. Operator --> evaluate the last two numbers, then push result again onto the stack

After this we do parenthesis example (polish notation)

--> push operand elements onto the stack
--> pop all the elements if operator, apply the operation and push the new element onto the stack again


```python
def postfix_expression_evaluation_robust(tokens):
    stack = [] # Local stack for the function
    for token in tokens:
        if token.isdigit():
            stack.append(int(token))
        else:
            op2 = stack.pop()
            op1 = stack.pop()
            if token == '+':
                stack.append(op1 + op2)
            elif token == '-':
                stack.append(op1 - op2)
            elif token == '*':
                stack.append(op1 * op2)
            elif token == '/':
                stack.append(op1 / op2)
    return stack.pop()
```

```python
tokens = "23*4+"
postfix_expression_evaluation_robust(tokens)
```

- For prefix expressions, also known as Polish notation, the operators come before their operands. When evaluating them using a stack, we typically process the expression from right to left.
