```python
n = int(input())
i = 1

while i <= n:
    # for the ith row, print i stars
    j = 1
    while j <= i:
        print("*", end = " ") # instead of going to the next line after print
        j = j + 1
    i = i + 1
    print() ## for new line
```

```python
# can also do this with for loop
n = int(input())
i = 1
for i in range(1,n+1):
    for j in range(1,i+1):
        print("*", end = " ")
    print()
        
```

```python
n = int(input()) ## n = 6, but output is 5
i = 1
for i in range(n):
    for _ in range(i):
        print("*", end = " ")
    print()
```
